```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

colordata = {}
for filename, colorname in [('colors_black.txt', 'k'), ('colors_green.txt', 'g'), ('colors_white.txt', 'white'), ('colors_semmi.txt', 'gray'), ('colors_red.txt', 'r')]:
    colordata[colorname] = np.loadtxt(filename)
x = np.linspace(0, 255, 1000)

```

```python
fig, ax = plt.subplots(1, 1)
for colorname, data in colordata.items():
    ax.plot(data[:, 0], data[:, 1], 'o', markerfacecolor=colorname, markeredgecolor='k')
ax.axis(xmin=0, xmax=255, ymin=0, ymax=255)
ax.plot(x, x, '--')
ax.set_xlabel('Red')
ax.set_ylabel('Green')
```

```python
fig, ax = plt.subplots(1, 1)
for colorname, data in colordata.items():
    ax.plot(data[:, 2], data[:, 1], 'o', markerfacecolor=colorname, markeredgecolor='k')
ax.axis(xmin=0, xmax=255, ymin=0, ymax=255)
ax.plot(x, x, '--')
ax.set_xlabel('Blue')
ax.set_ylabel('Green')
```
