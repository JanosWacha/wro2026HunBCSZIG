```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('run_log.txt')
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)
ax.plot(data[:, 0], data[:, 1], label='Front')
ax.plot(data[:, 0], data[:, 2], label='Left')
ax.plot(data[:, 0], data[:, 3], label='Right')
ax.plot(data[:, 0], data[:, 4], label='Back')
ax.legend(loc='best')
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)
ax.plot(data[:, 0], data[:, 1], label='Front+Back')
ax.legend(loc='best')
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)
ax.plot(data[:, 1], data[:, 4], '.', label='Front+Back')
ax.legend(loc='best')
```
